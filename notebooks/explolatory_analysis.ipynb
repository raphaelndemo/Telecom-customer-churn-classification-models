```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading the Dataset

df = pd.read_csv('../Data/bigml_59c28831336c6604c800002a.csv')
```

```python
sns.countplot(x=df["churn"])
plt.title("Churn Distribution")
plt.show()
```

The dataset shows significantly more customers staying than churning. Although churners represent a smaller portion of the dataset, they account for a disproportionate share of revenue risk. This highlights the importance of focusing on churn detection rather than relying solely on overall accuracy.

```python
sns.boxplot(x="churn", y="customer service calls", data=df)
plt.show()
```

Customers who churn tend to make more customer service calls compared to those who remain. This suggests that increased service interactions may reflect dissatisfaction or unresolved issues, making it a strong early indicator of churn risk.

```python
sns.countplot(x="international plan", hue="churn", data=df)
plt.show()
```

Customers subscribed to the international plan exhibit a higher churn rate compared to those without it. This indicates that certain plan types may be associated with higher dissatisfaction or pricing sensitivity, presenting an opportunity for targeted retention strategies within specific customer segments.

```python
df.groupby("churn").mean()
```

Churners generally show higher usage metrics and service interactions, reinforcing the idea that churn behavior is associated with measurable customer patterns rather than random events.
